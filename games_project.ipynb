```python
plt.figure(figsize=(30,25))
plt.subplot(331)
sns.barplot(x='Year' , y= 'Global_Sales' , data=game_df ) 
plt.xticks(rotation=50)
plt.title("Golbal Sales of Video Games in past Years ") 
plt.subplot(332)
sns.barplot(x='Year' , y= 'NA_Sales' , data=game_df ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games in North Amercia in past Years ") 
plt.subplot(333)
sns.barplot(x='Year' , y= 'JP_Sales' , data=game_df ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games in Japan in past Years ") 
plt.subplot(334)
sns.barplot(x='Year' , y= 'EU_Sales' , data=game_df ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games in Europe in past Years ") 
plt.subplot(335)
sns.barplot(x='Year' , y= 'Other_Sales' , data=game_df ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games in Other Countries in past Years ") 

plt.show()
```

### The above graphs show the sales for each country according to the years.
Sales in all countries hit their all time high in the decade 1980-90 and have since seen a decline in sales.
In the later years, only Europe and Other Countries show an increasing trend in sales.

```python
plt.figure(figsize=(35,30))
plt.subplot(331)
sns.boxplot(x = 'Genre', y = 'Global_Sales' , data = game_df,palette="Set2",fliersize = 0)
plt.xticks(rotation=40)
plt.title("Global Sales of Video Games of different Genre")
plt.ylim(0, 2)

plt.subplot(332)
sns.boxplot(x = 'Genre', y = 'NA_Sales' , data = game_df,palette="Set2",fliersize = 0)
plt.xticks(rotation=40)
plt.title("Sales of Video Games of different Genre in North America ") 
plt.ylim(0, 2)

plt.subplot(333)
plt.ylim(0, 1)
sns.boxplot(x = 'Genre', y = 'EU_Sales' , data = game_df,palette="Set2",fliersize = 0)
plt.xticks(rotation=40)
plt.title("Sales of Video Games of different Genre in Europe ")

plt.subplot(334)
plt.ylim(0, 1)
sns.boxplot(x = 'Genre', y = 'JP_Sales' , data = game_df,palette="Set2",fliersize = 0)
plt.xticks(rotation=40)
plt.title("Sales of Video Games of different Genre in Japan ")

plt.subplot(335)
sns.boxplot(x = 'Genre', y = 'Other_Sales' , data = game_df,palette="Set2",fliersize = 0)
plt.xticks(rotation=40)
plt.title("Sales of Video Games of different Genre in Other Countries ")
plt.ylim(0,1)
plt.show()
```

In Global sales, as well as countries like NA and Europe, platform and shooter sales are the most popular.
In Japan, Role Playing games are in the lead, followed by fighting and strategy. Sales of racing games are close to 0.
Other countries record very low sales across all categories, shooter games in the lead.

```python
plt.figure(figsize=(28,23))
plt.subplot(331)
sns.barplot(x='Platform' , y= 'Global_Sales' , data=game_df,palette="cool" ) 
plt.xticks(rotation=50)
plt.title("Global Sales of Video Games of different Platform")
plt.subplot(332)
sns.barplot(x='Platform' , y= 'NA_Sales' , data=game_df,palette="cool") 
plt.xticks(rotation=50)
plt.title("Sales of Video Games of different Platform in North America ") 
plt.subplot(333)
sns.barplot(x='Platform' , y= 'JP_Sales' , data=game_df,palette="cool") 
plt.xticks(rotation=50)
plt.title("Sales of Video Games of different Platform in Japan ")
plt.subplot(334)
sns.barplot(x='Platform' , y= 'EU_Sales' , data=game_df,palette="cool" ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games of different Platform in Europe ")
plt.subplot(335)
sns.barplot(x='Platform' , y= 'Other_Sales' , data=game_df,palette="cool" ) 
plt.xticks(rotation=50)
plt.title("Sales of Video Games of different Platform in Other Countries ")
plt.show()
```

These graphs show the platform wise sales in different countries. 
Games by NES and GB are the most purchased, followed by GEN.
Platforms like WS, NG, TG16, 3DO, GG, PCFX are being purchased only in Japan.
GEN and 2600 are at second and third position in North America.
SNES and WS follow NES and GB in Japan. 2600, which has the third highest sales in North America, recorded 0 sales in Japan.
In Europe, GB is followed by PS4, GEN. amd NES.
In countries other than the ones mentioned, PS4 ranks highest in sales, followed by GB and PS3.

```python
publisher_df = game_df[['Global_Sales','Publisher']].groupby('Publisher').mean().sort_values('Global_Sales').tail(50)
publisher_df.plot(kind = 'bar' ,title = 'Average  Video Games Sales Over different Publishers', figsize = (20,5), width = 0.9)
plt.show()
publisher_df = game_df[['JP_Sales','Publisher']].groupby('Publisher').mean().sort_values('JP_Sales').tail(50)
publisher_df.plot(kind = 'bar' ,title = 'Average  Video Games Sales Over different Publishers', figsize = (20,5), width = 0.9)
plt.show()
publisher_df = game_df[['NA_Sales','Publisher']].groupby('Publisher').mean().sort_values('NA_Sales').tail(50)
publisher_df.plot(kind = 'bar' ,title = 'Average  Video Games Sales Over different Publishers', figsize = (20,5), width = 0.9)
plt.show()
publisher_df = game_df[['EU_Sales','Publisher']].groupby('Publisher').mean().sort_values('EU_Sales').tail(50)
publisher_df.plot(kind = 'bar' ,title = 'Average  Video Games Sales Over different Publishers', figsize = (20,5), width = 0.9)
plt.show()
publisher_df = game_df[['Other_Sales','Publisher']].groupby('Publisher').mean().sort_values('Other_Sales').tail(50)
publisher_df.plot(kind = 'bar' ,title = 'Average  Video Games Sales Over different Publishers', figsize = (20,5), width = 0.9)
plt.show()
```

These graphs show the mean sales of each publisher, considering sales from year 1980 to 2020.
-->Palcom has recorded the highest sales in GLobal Market and majorly in North America
-->Enix Corporation has recorded the highest sales in Japan
-->Red ORB Corporation has recorded the highest sales in Europe 
-->Hello Games Corporation has recorded the highest sales in Other Countries
